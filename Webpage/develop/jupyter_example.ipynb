## Writing admonitions

You can write admonitions in html code. Unfortunately it has to be this way... 

You need to write as the class inside the `<div>` tag "admonition {type}", e.g. "admonition note" or "admonition tip".
Then in the following `<p>` block, you should add the class "admonition-title", followed by the title of the admonition.
In the next `<p>` block, you can add the content.

<div class="admonition note">
    <p class="admonition-title">Title of the admonition</p>
    <p> &#128368; <strong>Time Estimation:</strong> X minutes<br>
        &#128172; <strong>Learning Objectives:</strong><br>
        &emsp;1. First item<br>
        &emsp;2. Second item<br>
   </p>
</div>

Here is an example of a dropdown admonition. 
You do not need to provide the class in this case since it is wrapped inside the `<details>` tag,
but you need to provide a suitable admonition type in class (e.g., "warning" or "note"). 
The title of the admonition is inside the `<summary>` tag.

<details class="warning">
    <summary>Title of drop-down</summary>
    <p>
    Content
    </p>
</details>

Inside the html 

**Bold text**
You will need `<strong>` and `</strong>` tags for bold text 

*Italic text*
You will need the `<i>` and `</i>` tags for italic text

Tab
You will need the tag `&emsp;` to add a 4 space tab in html code
